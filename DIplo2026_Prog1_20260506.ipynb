```python
# Signo numeral # sirve para comentar el código, nuestro programa no hace nada
#Los comentarios sirven para indicar y


                 #Variables: son espacios de memoria destinados a almacenar información
variable1 = 3    #reservar un espacio de memoria que le voy a dar el nombre variable1, y le voy a dar (=) el valor 3
print(variable1)
variable1 = -99  #pisa la variable anterior
print(variable1) #print voy a mostar por pantalla el contenido de la variable1

type(variable1)  #type = "tipo" in Spanish, o sea que me muestra de qué tipo es variable1, el resultado de la variable es de tipo int
                 # dió de resultado int que significa que es un número entero, o sea los que van desde el -infinito al +infinito creciendo/decreciendo de 1 en 1 sin decimales

variable2 = 9.3
print(variable2)
type(variable2)  #el tipo float significa flotante, son los números racionales, positivos y negativos con decimales (se pueden convertir a fracción)

variable3 = "Hola"
print(variable3) #Variable STRING (cadena de caracteres)
type(variable3)

variable4 = True   #Variables booleanas TRUE == Verdadero  FALSE == Falso
variable5 = False  #Variables booleanas FALSE == Falso
type(variable4)
variable6 = "true"
type(variable6)


#¡OJO! Lenguaje de bajo tipado ¡A una variable le puedo cambiar el tipo de dato!
print(type(variable1))
variable1 = "Ahora soy string"
print(type(variable1))
variable1 = -4.5
print(type(variable1))

#IMPORTANTE: El nombre de la variable tiene que ser significativo
#RECOMENDACIÓN: En la medida de lo posible NO cambiar el tipo de dato de las variables (int <==> float)

#CONSTANTES (no debrían modificar su valor)  ¡CUIDADO! EN Pyton son variables que debo mantener sin alteración
#Definir constantes todas en mayúcula
CANTIDAD_PERSONAS = 45
print(CANTIDAD_PERSONAS)
```
